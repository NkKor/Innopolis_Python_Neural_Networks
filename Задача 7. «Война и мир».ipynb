```python
# Задача 7. «Война и мир»

def word_finder(book, words):

    found_lines = []

    file = book.split("\n")
    lines = len(file)
    print("Строк - ", lines)
    for line in file:
      for word in words:
        if word in line:
          found_lines.append(line.strip())
          break
    print(f"Строк с упоминанием {words} -  {len(found_lines)}")
    for line in found_lines:
        print(line)
    return f"""Предложения со словом «Андрей»:
    {found_lines}"""

def statistics(full_text):
    return f"""Статистика
Всего символов во фрагменте: {len(full_text)}
а : {full_text.count("а")/len(full_text):.3f}
б : {full_text.count("б")/len(full_text):.3f}
в : {full_text.count("в")/len(full_text):.3f}
. : {full_text.count(".")/len(full_text):.3f}
! : {full_text.count("!")/len(full_text):.3f}
"""

with open('war_and_peace.txt', 'r') as file:
    book = file.read()
    words = ["Андрей", "Андрея", "Андрею", "Андреем", "Андрее"]
    print(statistics(book))
    print(word_finder(file,words))
```
